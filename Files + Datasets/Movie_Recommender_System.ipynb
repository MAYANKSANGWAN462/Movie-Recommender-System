```python
# now we have to convert this : '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# to this : ['Action','Adventure','Fantasy','SciFi']
```

```python
def convert(obj):
  L = []
  for i in obj:
    L.append(i['name'])
  return L
```

```python
#helped to convert the string of list ----to---> list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert) # apply(convert) automatically passes each cellâ€™s value from the column to the convert() function â€” you donâ€™t need to give the argument manually.
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
#now we are removing the spcae from the genres in order to make the person name the single entity
# # 1ï¸âƒ£ List Comprehension
# squares = [x**2 for x in range(5)]                     # [0, 1, 4, 9, 16]

# # 2ï¸âƒ£ Dictionary Comprehension
# square_dict = {x: x**2 for x in range(3)}              # {0: 0, 1: 1, 2: 4}

# # 3ï¸âƒ£ Set Comprehension
# unique_mod = {x % 3 for x in [1, 2, 3, 4, 5]}          # {0, 1, 2}

# # 4ï¸âƒ£ Generator Expression
# gen = (x**2 for x in range(3))                         # generator -> next(gen) gives 0, 1, 4

# # 5ï¸âƒ£ Lambda Function
# add = lambda a, b: a + b                               # add(2, 3) -> 5

# # 6ï¸âƒ£ Conditional (Ternary) Expression
# result = "Even" if 4 % 2 == 0 else "Odd"              # "Even"

# # 7ï¸âƒ£ any() and all() with Comprehension
# nums = [2, 4, 6]
# all_even = all(n % 2 == 0 for n in nums)               # True
# any_odd = any(n % 2 != 0 for n in nums)                # False

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split(): #here the text.split() is making it to the list from string
    y.append(ps.stem(i))

  return " ".join(y)

```

```python
#now its the movie recommending function
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  #now for that particular movie we want its similarity coefficients with other movies and also we are having the movie index and need the cosine similarity with respect to other
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]



  for i in movies_list:
    print(new_df.iloc[i[0]].title)
    print(new_df.iloc[i[0]])
    print(i[0])



```
